# Literature Review Entry

## 1. Bibliographic entry

Campbell, John Y., and Samuel B. Thompson. 2008. "Predicting Excess Stock Returns Out of Sample: Can Anything Beat the Historical Average?" *The Review of Financial Studies* 21 (4): 1509–1531. https://doi.org/10.1093/rfs/hhm055

## 2. Thesis

Imposing weak, theory-motivated restrictions on predictive regressions — forcing regression coefficients to carry their economically expected sign and truncating the equity-premium forecast at zero — lets valuation ratios and other standard predictors beat the naive historical-average benchmark out of sample, overturning the pessimistic verdict of Goyal and Welch.

## 3. Main points and findings

Responding to Goyal and Welch's evidence that most predictors fail to beat the prevailing historical mean out of sample, Campbell and Thompson add two simple economic constraints: (i) set a predictor's regression slope to zero whenever it takes the "wrong" sign relative to theory, and (ii) replace any negative fitted equity premium with zero, since a rational investor would not expect a negative excess return. These restrictions shrink forecast variance and consistently push the out-of-sample R-squared into positive territory for many predictors (dividend- and earnings-price ratios, the T-bill rate, and others), whereas the same predictors are unreliable when left unrestricted. The gains in out-of-sample R-squared are small in absolute terms — on the order of half a percent per month — but the authors show through a mean-variance argument that even such small R-squared values are economically meaningful, because out-of-sample R-squared should be judged against the squared Sharpe ratio rather than against zero. The paper thus reframes "small statistical predictability" as "material economic value" for a real-time investor.

## 4. Relevance to the current research project

This paper supplies the exact economic-evaluation machinery that Gu, Kelly, and Xiu (2020) — our main replication target — use to translate predictive out-of-sample R-squared into Sharpe-ratio improvements and constrained market-timing strategies, so replicating Campbell and Thompson's R-squared-to-Sharpe mapping and their positivity/leverage restrictions is a prerequisite for reproducing our main paper's portfolio-level economic results.


# Key Contributions and Methodology

## Setup and benchmark

The object of interest is the one-period-ahead equity premium (excess market return). The competing forecasts are a predictive regression versus the *prevailing historical average*:

$$
\hat r^{HA}_{t+1} = \frac{1}{t}\sum_{s=1}^{t} r_s , \qquad
\hat r^{REG}_{t+1} = \hat\alpha_t + \hat\beta_t\, x_t ,
$$

where $x_t$ is a predictor (for example the dividend-price or earnings-price ratio). Forecasts are strictly out of sample: parameters at date $t$ use only data through $t$. Performance is measured with the out-of-sample $R^2$ relative to the historical mean,

$$
R^2_{oos} = 1 - \frac{\sum_t \left(r_{t+1}-\hat r^{REG}_{t+1}\right)^2}{\sum_t \left(r_{t+1}-\hat r^{HA}_{t+1}\right)^2}.
$$

Goyal and Welch (2008) documented that this quantity is typically negative — the historical average wins.

## The two economic restrictions

1. **Sign restriction on the slope.** If the estimated coefficient $\hat\beta_t$ has the theoretically wrong sign (for example a negative loading on the dividend yield), set the slope to zero and fall back to the historical average for that month. Equivalently, admit the regression forecast only when the coefficient sign matches theory.

2. **Positivity restriction on the forecast.** If the fitted premium is negative, replace it with zero:

$$
\hat r^{\,+}_{t+1} = \max\!\left(0,\; \hat r^{REG}_{t+1}\right).
$$

Both restrictions inject prior economic knowledge, reduce estimation-driven forecast variance, and are what turn a losing unrestricted regression into one that beats the historical average.

## From statistical R-squared to economic value

The paper's most reused result links predictability to portfolio performance for a mean-variance investor. A useful benchmark is that the out-of-sample $R^2$ should be compared with the squared (per-period) Sharpe ratio $S^2$ of the market, not with zero: when $R^2_{oos}$ is of the same order as $S^2$, the economic gains from timing are already substantial. Gu, Kelly, and Xiu (2020) express the same idea through the achievable Sharpe ratio of a timing strategy,

$$
SR^{*} = \sqrt{\frac{SR^2 + R^2}{1 - R^2}},
$$

where $SR$ is the buy-and-hold Sharpe ratio and $R^2$ is the predictive out-of-sample $R^2$. This is the bridge from a small $R^2$ to a meaningful improvement in risk-adjusted return.


# Relation to the Main Paper (Gu, Kelly & Xiu, 2020)

Campbell and Thompson (2008) is a methodological ancestor of the economic-evaluation section of our main paper. Three concrete links matter for the replication:

1. **R-squared-to-Sharpe translation.** Gu, Kelly, and Xiu use the Campbell–Thompson $SR^{*}$ formula above to convert each model's portfolio-level out-of-sample $R^2$ (their Table 5) into an annualized Sharpe-ratio gain (their Internet Appendix Table A.7). Our replication reuses this exact mapping.

2. **Constrained market-timing strategy.** Gu, Kelly, and Xiu "follow Campbell and Thompson (2008) exactly" when building their timing strategy: scale the position up or down as the expected return rises or falls, impose a maximum leverage constraint of 50%, and forbid short sales for long-only portfolios. This is the trading rule behind their reported S&P 500 timing Sharpe ratio of 0.77 for NN3 versus 0.51 buy-and-hold.

3. **Disciplined benchmarking.** Campbell and Thompson's argument that a raw historical mean is a noisy, easily beaten benchmark motivates Gu, Kelly, and Xiu's decision to benchmark *individual-stock* out-of-sample $R^2$ against a forecast of zero rather than the historical average — a subtle but important design choice in our replication's evaluation metric.

**Takeaway for our project:** when we reproduce the portfolio and Sharpe-ratio results, the positivity/leverage restrictions and the $R^2$-to-Sharpe formula come from this paper, so it should be cited wherever the replication moves from predictive accuracy to economic performance.
